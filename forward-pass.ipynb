```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SimpleNet:
    '''A simple neural network with 2 inputs and 1 output'''
    def __init__(self, weights, bias=1):
        self.weights = weights
        self.bias = bias
    
    def _sigmoid(self, z):
        '''Sigmoid activation function'''
        return 1 / (1 + np.exp(-z))

    def calculate(self, inputs):
        '''Calculate the output of the network'''
        return self._sigmoid(np.append(np.array([self.bias]), inputs) @ self.weights)
    
    def graph(self, inputs: list, table=False):
        '''Draw the results for a given set of inputs'''
        data = pd.DataFrame({'in': [x[0] for x in inputs], 'out': [self.calculate(x) for x in inputs], 'Inputs': [str(x) for x in inputs]})
        # sns.scatterplot(data=data, x='x', y='y', hue='Inputs', alpha=0.7)
        if table:
            for d in data.values:
                print(f'Inputs: {d[2]}, Output: {d[1]:.5f} [~{round(d[1])}]')
        sns.pointplot(data=data, x='in', y='out', hue='Inputs', dodge=True)
```

```python
# AND gate
and_net = SimpleNet(weights=np.array([-30, 20, 20]))
and_net.graph(inputs=list([np.array([0, 0]), np.array([0, 1]), np.array([1, 0]), np.array([1, 1])]), table=True)
```

```python
# NOT AND NOT gate
nan_net = SimpleNet(weights=np.array([10, -20, -20]))
nan_net.graph(inputs=list([np.array([0, 0]), np.array([0, 1]), np.array([1, 0]), np.array([1, 1])]), table=True)
```

```python
# OR gate
nan_net = SimpleNet(weights=np.array([-10, 20, 20]))
nan_net.graph(inputs=list([np.array([0, 0]), np.array([0, 1]), np.array([1, 0]), np.array([1, 1])]), table=True)
```
