```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('data/exam-iq.csv')

sns.scatterplot(data=data, y='IQ', x='Hours', hue='Pass')
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics

logReg = LogisticRegression()
logReg.fit(data[['Hours', 'IQ']], data['Pass'])
logReg.predict(data[['Hours', 'IQ']])

accuracy = logReg.score(data[['Hours', 'IQ']], data['Pass'])
precision = metrics.precision_score(data['Pass'], logReg.predict(data[['Hours', 'IQ']]))
recall = metrics.recall_score(data['Pass'], logReg.predict(data[['Hours', 'IQ']]))
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
```

```python
sns.scatterplot(data=data, y='IQ', x='Hours', hue='Pass')
sns.lineplot(x=data['Hours'].to_numpy(), y=-logReg.coef_[0][0]/logReg.coef_[0][1]*data['Hours'] - logReg.intercept_/logReg.coef_[0][1], label='Decision Boundary')
```
